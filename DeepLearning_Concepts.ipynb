## Different Types of Pooling

```python
from skimage.measure import block_reduce
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
img = imread('AI_icon')
```

```python
mean_pool = block_reduce(img, block_size=(9,9,1), func=np.mean)
max_pool = block_reduce(img, block_size=(9,9,1), func=np.max)
min_pool = block_reduce(img, block_size=(9,9,1), func=np.min)
```

```python
plt.figure(1)
plt.subplot(221)
imgplot = plt.imshow(img)
plt.title('Original Image', fontsize=8)
# plt.show()

plt.subplot(222)
imgplot1 = plt.imshow(mean_pool)
plt.title("Average Pooling", fontsize=8)
# plt.show()

plt.subplot(223)
imgplot1 = plt.imshow(max_pool)
plt.title("Max Pooling", fontsize=8)
# plt.show()

plt.subplot(224)
imgplot1 = plt.imshow(min_pool)
plt.title("Min Pooling", fontsize=8)
plt.show()
```

Average pooling smooths out the image. Sharp features may not be highlighted.

Max pooling highlights bright pixles of the images. Helpful when the background of the image is dark and ligher pixels are of interest

Min pooing is reverse of max pooling. Helpful when there is a need to highlight dark pixels. Helpful when the background of the image is lighter.
